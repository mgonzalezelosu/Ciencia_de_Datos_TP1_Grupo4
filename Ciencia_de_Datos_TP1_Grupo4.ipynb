```python
#variables de interÃ©s
vars_corr = ["CH04","CH06","CH07","CH08","NIVEL_ED","ESTADO","CAT_INAC","IPCF"]
#funciÃ³n para preparar datos por aÃ±o ---
def preparar_corr(df, nombre=""):
    # Filtrar columnas
    df_corr = df[vars_corr].copy()

#hago las dummies para las variables categÃ³ricas
    df_corr = pd.get_dummies(
        df_corr,
        columns=["CH04", "CH07", "CH08", "ESTADO", "CAT_INAC"],  #sexo, cobertura, estado laboral, categorÃ­a inactividad
        drop_first=True  #evitar colinealidad perfecta
    )

    # Renombrar algunas columnas para que tengan sentido
    renombres = {}
    if "CH04_2" in df_corr.columns:
        renombres["CH04_2"] = "Mujer"
    # Cobertura mÃ©dica
    for col in df_corr.columns:
        if col.startswith("CH08_"):
            renombres[col] = "CobMed_" + col.split("_")[1]
    # Estado
    for col in df_corr.columns:
        if col.startswith("ESTADO_"):
            renombres[col] = "Estado_" + col.split("_")[1]
    # CategorÃ­a inactividad
    for col in df_corr.columns:
        if col.startswith("CAT_INAC_"):
            renombres[col] = "Inac_" + col.split("_")[1]

    df_corr = df_corr.rename(columns=renombres)

    return df_corr

# --- Preparar para 2005 y 2025 ---
df_2005_corr = preparar_corr(data_2005, nombre="2005")
df_2025_corr = preparar_corr(data_2025, nombre="2025")

# --- Graficar ---
plt.figure(figsize=(12,8))
sns.heatmap(df_2005_corr.corr(), vmin=-1, vmax=1, annot=False, cmap="coolwarm")
plt.title("Matriz de correlaciÃ³n - RegiÃ³n 42 (2005)")
plt.show()

plt.figure(figsize=(12,8))
sns.heatmap(df_2025_corr.corr(), vmin=-1, vmax=1, annot=False, cmap="coolwarm")
plt.title("Matriz de correlaciÃ³n - RegiÃ³n 42 (2025)")
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- Config base (la tuya) ---
vars_corr = ["CH04","CH06","CH07","CH08","NIVEL_ED","ESTADO","CAT_INAC","IPCF"]

def preparar_corr(df):
    base = df[vars_corr].copy()
    # continuas a numÃ©rico
    for c in ["CH06","NIVEL_ED","IPCF"]:
        base[c] = pd.to_numeric(base[c], errors="coerce")
    # dummies (drop_first=True como pediste)
    base = pd.get_dummies(base,
                          columns=["CH04","CH07","CH08","ESTADO","CAT_INAC"],
                          drop_first=True, dtype="int")
    return base

label_map = {
    "CH06":"Edad (aÃ±os)","NIVEL_ED":"Nivel educativo (1â€“7)","IPCF":"Ingreso per cÃ¡pita familiar",
    "CH04_2":"Mujer (1=sÃ­)",
    "CH07_2":"Casado/a","CH07_3":"Separado/a o divorciado/a","CH07_4":"Viudo/a","CH07_5":"Soltero/a",
    "CH08_2":"Mutual/prepaga","CH08_3":"Planes/seguros pÃºblicos","CH08_4":"No paga / no descuentan",
    "CH08_9":"Ns/Nr","CH08_12":"OS + prepaga","CH08_13":"OS + planes",
    "CH08_23":"Prepaga + planes","CH08_123":"OS + prepaga + planes",
    "ESTADO_1":"Ocupado/a","ESTADO_2":"Desocupado/a","ESTADO_3":"Inactivo/a","ESTADO_4":"Menor de 10 aÃ±os",
    "CAT_INAC_1":"Jubilado/a â€“ pensionado/a","CAT_INAC_2":"Rentista","CAT_INAC_3":"Estudiante",
    "CAT_INAC_4":"Ama de casa","CAT_INAC_5":"Menor de 6 aÃ±os","CAT_INAC_6":"Discapacitado/a","CAT_INAC_7":"Otros inactivos",
}

def orden_variables(df_corr):
    bloques = [
        ["CH06","NIVEL_ED","IPCF"],
        ["CH04_2"],
        ["CH07_2","CH07_3","CH07_4","CH07_5"],
        ["CH08_2","CH08_3","CH08_4","CH08_9","CH08_12","CH08_13","CH08_23","CH08_123"],
        ["ESTADO_1","ESTADO_2","ESTADO_3","ESTADO_4"],
        ["CAT_INAC_1","CAT_INAC_2","CAT_INAC_3","CAT_INAC_4","CAT_INAC_5","CAT_INAC_6","CAT_INAC_7"],
    ]
    return [v for g in bloques for v in g if v in df_corr.columns]

def plot_corr(df_corr, year, region="Cuyo", umbral=0.20):
    # --- MISMA MUESTRA PARA TODA LA MATRIZ (listwise deletion) ---
    df_list = df_corr.dropna(subset=df_corr.columns.tolist()).copy()

    # --- Matriz ordenada ---
    orden = orden_variables(df_list)
    corr = df_list[orden].corr(numeric_only=True)
    corr.index = [label_map.get(c,c) for c in corr.index]
    corr.columns = [label_map.get(c,c) for c in corr.columns]

    # --- MÃ¡scara triÃ¡ngulo superior ---
    mask = np.triu(np.ones_like(corr, dtype=bool))

    # --- Anotaciones: solo |r| >= umbral; el resto vacÃ­o (sin 'nan') ---
    txt = corr.where(corr.abs() >= umbral).round(2).astype(str)
    txt = txt.where(~txt.isin(["nan","NaN"]), "")  # vaciar textos no relevantes

    # --- Plot minimalista (sin lÃ­neas, sin bordes) ---
    plt.figure(figsize=(14,11))
    ax = sns.heatmap(
        corr, mask=mask, cmap="RdBu_r", center=0, vmin=-1, vmax=1,
        annot=txt, fmt="", annot_kws={"size":8},
        linewidths=0, linecolor=None,
        square=False,
        cbar_kws={"shrink": .8, "label": "Coef. de correlaciÃ³n (Pearson)"}
    )
    ax.set_title(f"Matriz de correlaciÃ³n â€“ {region} ({year})",
                 fontsize=16, fontweight="bold", pad=12)
    plt.xticks(rotation=45, ha="right", fontsize=9)
    plt.yticks(fontsize=9)
    plt.tight_layout()
    plt.show()

# === Ejecutar con tus dataframes ===
df_2005_corr = preparar_corr(data_2005)
df_2025_corr = preparar_corr(data_2025)

plot_corr(df_2005_corr, year="2005", region="Cuyo", umbral=0.20)
plot_corr(df_2025_corr, year="2025", region="Cuyo", umbral=0.20)
```

```python
tabla_equiv.columns = ["EDAD", "MUJERES", "VARONES"]

#FunciÃ³n para mapear edad + sexo a adulto_equiv
def asignar_adulto_equiv(row, tabla=tabla_equiv):
    edad = row["CH06"]
    sexo = row["CH04"]  # 1 = hombre, 2 = mujer

    if sexo == 1:  # hombre â†’ usar columna VARONES
        valor = tabla.loc[tabla["EDAD"] == edad, "VARONES"]
    elif sexo == 2:  # mujer â†’ usar columna MUJERES
        valor = tabla.loc[tabla["EDAD"] == edad, "MUJERES"]
    else:
        return None  # si CH04 no es 0 o 1

    # Si no encuentra edad en la tabla
    if valor.empty:
        return None
    return valor.values[0]

# ðŸ“Œ Crear la nueva columna en base_2005
data_final["adulto_equiv"] = data_final.apply(asignar_adulto_equiv, axis=1)
data_2025['adulto_equiv'] = data_2025.apply(asignar_adulto_equiv, axis=1)
data_2005['adulto_equiv'] = data_2005.apply(asignar_adulto_equiv, axis=1)


data_final[["CH04", "CH06", "adulto_equiv"]].head()
```

```python
#MATCHING CODOSU y SUMANDO VALORES de c/u en columna adulto_equiv
data_final["ad_equiv_hogar"] = (
    data_final.groupby("CODUSU")["adulto_equiv"].transform("sum")
)

data_final[["CODUSU", "CH04", "CH06", "adulto_equiv", "ad_equiv_hogar"]].head(20)

data_2005["ad_equiv_hogar"] = (
    data_2005.groupby("CODUSU")["adulto_equiv"].transform("sum")
)

data_2005[["CODUSU", "CH04", "CH06", "adulto_equiv", "ad_equiv_hogar"]].head(20)

data_2025["ad_equiv_hogar"] = (
    data_2025.groupby("CODUSU")["adulto_equiv"].transform("sum")
)

data_2025[["CODUSU", "CH04", "CH06", "adulto_equiv", "ad_equiv_hogar"]].head(20)
```

```python
# FunciÃ³n de resumen para variable booleana
def resumen_booleana(df, col="pobre"):
    n = len(df)
    s = df[col].sum()             # cuenta los True
    n0 = n - s                    # cuenta los False
    p = (s / n * 100) if n > 0 else float("nan")
    return pd.Series({
        "Observaciones": n,
        "Pobres (True)": s,
        "No pobres (False)": n0,
        "% Pobres": round(p, 2),
        "% No pobres": round(100 - p, 2) if pd.notna(p) else float("nan")
    })

# Tabla comparativa
tabla_2005 = resumen_booleana(respondieron05)
tabla_2025 = resumen_booleana(respondieron25)

tabla_pobre = pd.DataFrame([tabla_2005, tabla_2025], index=["2005", "2025"])
display(tabla_pobre)
```

```python
def resumen_pobre(df, year):
    total = len(df)
    pobres = df["pobre"].sum()
    no_pobres = total - pobres
    pct_pobres = (pobres / total * 100) if total > 0 else float("nan")
    return pd.Series({
        "AÃ±o": year,
        "Observaciones": total,
        "Pobres": int(pobres),
        "No pobres": int(no_pobres),
        "% Pobres": round(pct_pobres, 2),
        "% No pobres": round(100 - pct_pobres, 2),
        "DesvÃ­o estÃ¡ndar": round(df["pobre"].std(), 3)
    })

tabla_2005 = resumen_pobre(respondieron05, 2005)
tabla_2025 = resumen_pobre(respondieron25, 2025)

#esta tabla es parte del punto 9 tambien CREO
tabla_pobre = pd.DataFrame([tabla_2005, tabla_2025]).set_index("AÃ±o")
display(tabla_pobre)
```
